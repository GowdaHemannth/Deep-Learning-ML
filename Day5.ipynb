```python
## The techiques to tackle the Overfitting and Underfitting are called Regurilization Techniques 

## By Droppig the Neurons We can reduce the Overfitting by Decreas g the Number of nerons 
import pandas as pd
df=pd.read_csv("sonar_dataset.csv",header=None) ## SInce these Daatset does not Conatin ANy Columns So 
df.head()
```

```python
##
x = df.drop(60, axis=1)  # All columns except last
y = df[60]
y.head()
```

```python
y=pd.get_dummies(y,drop_first=True).astype(int)
y.head()
```

```python
## Dived into Tarina nd Test thing 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
X_train.head()
```

```python
## Now Building the Neurol Networks
import tensorflow as tf
from tensorflow import keras

model=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
    
    
])

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```
